# Data matrix creation

## Read individual files

```python
import pandas as pd


demographics =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/demographics_icd.csv')
vitals_min =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/vitals_min_icd_standardised.csv')
vitals_max =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/vitals_max_icd_standardised.csv')
vitals_avg =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/vitals_avg_icd_standardised.csv')
vitals_stddev =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/vitals_stddev_icd_standardised.csv')
vitals_first =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/vitals_first_icd_standardised.csv')
vitals_last =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/vitals_last_icd_standardised.csv')
lab_measurements_min =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/lab_measurements_min_icd_standardised.csv')
lab_measurements_max =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/lab_measurements_max_icd_standardised.csv')
lab_measurements_avg =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/lab_measurements_avg_icd_standardised.csv')
lab_measurements_stddev =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/lab_measurements_stddev_icd_standardised.csv')
lab_measurements_first =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/lab_measurements_first_icd_standardised.csv')
lab_measurements_last =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/standardised/lab_measurements_last_icd_standardised.csv')
deaths =  pd.read_csv('/superbugai-data/yash/chapter_1/workspace/EHRQC/data/icd_cohort_test/deaths.csv')
```

## Merge all the files

```python
demographics['gender'] = demographics.gender.apply(lambda x: 1 if(x == 'M') else 0)
```

```python
demographics['ethnicity_WHITE'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'WHITE') else 0)
demographics['ethnicity_BLACK'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'BLACK/AFRICAN AMERICAN') else 0)
demographics['ethnicity_UNKNOWN'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'UNKNOWN') else 0)
demographics['ethnicity_OTHER'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'OTHER') else 0)
demographics['ethnicity_HISPANIC'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'HISPANIC/LATINO') else 0)
demographics['ethnicity_ASIAN'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'ASIAN') else 0)
demographics['ethnicity_UNABLE_TO_OBTAIN'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'UNABLE TO OBTAIN') else 0)
demographics['ethnicity_AMERICAN_INDIAN'] = demographics.ethnicity.apply(lambda x: 1 if(x == 'AMERICAN INDIAN/ALASKA NATIVE') else 0)

demographics.drop(columns='ethnicity', inplace=True)
```

```python
import pandas as pd


mergedDf = pd.merge(demographics, deaths, how='inner', on='person_id')
mergedDf = pd.merge(mergedDf, vitals_min, how='inner', on='person_id')
mergedDf = pd.merge(mergedDf, vitals_max, how='inner', on='person_id', suffixes=('_min', '_max'))
mergedDf = pd.merge(mergedDf, vitals_avg, how='inner', on='person_id')
mergedDf = pd.merge(mergedDf, vitals_stddev, how='inner', on='person_id', suffixes=('_avg', '_stddev'))
mergedDf = pd.merge(mergedDf, vitals_first, how='inner', on='person_id')
mergedDf = pd.merge(mergedDf, vitals_last, how='inner', on='person_id', suffixes=('_first', '_last'))
mergedDf = pd.merge(mergedDf, lab_measurements_min, how='inner', on='person_id')
mergedDf = pd.merge(mergedDf, lab_measurements_max, how='inner', on='person_id', suffixes=('_min', '_max'))
mergedDf = pd.merge(mergedDf, lab_measurements_avg.add_suffix('_avg'), how='inner', left_on='person_id', right_on='person_id_avg')
mergedDf = pd.merge(mergedDf, lab_measurements_stddev.add_suffix('_stddev'), how='inner', left_on='person_id', right_on='person_id_stddev')
mergedDf = pd.merge(mergedDf, lab_measurements_first, how='inner', on='person_id')
mergedDf = pd.merge(mergedDf, lab_measurements_last, how='inner', on='person_id', suffixes=('_first', '_last'))
mergedDf = mergedDf.drop(['person_id_avg', 'person_id_stddev'], axis = 1)
mergedDf
```
